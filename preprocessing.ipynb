Pearson correlation coefficient is a measure of this directionality 
score closer to 1 between pairs of features means that they move together in the same direction more strongly
score closer to 0 means features are not correlated and a score close to -1 means  they are more strongly negatively correlated, meaning one feature increases in value while the other decreases.
